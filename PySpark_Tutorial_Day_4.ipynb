```
File='/FileStore/tables/Department.csv'
df_pyspark = spark.read.csv(File ,header=True , inferSchema= True)
```

```
df_pyspark.show()
```

```
## using GroupBy with Name
df_pyspark.groupBy('Name').sum().show()
```

```
## finding the maximum salary of a department
df_pyspark.groupBy('Department').sum().show()
```

```
## Using count function
df_pyspark.groupBy('Department').count().show()
```

```
## finding the sum off all the salary
df_pyspark.agg({'Salary':'sum'}).show()
```

```
## finding the max salary
df_pyspark.agg({'Salary':'max'}).show()
```

```
## finding the min salary
df_pyspark.agg({'Salary':'min'}).show()
```

```
## finding the avg or mean salary
df_pyspark.agg({'Salary':'avg'}).show()
df_pyspark.agg({'Salary':'mean'}).show()
```

```
## getting the minimum salary of a department
df_pyspark.groupBy('Department').min().show()
```

```
## getting the max salary in a department
df_pyspark.groupBy('Department').max().show()
```

```
from pyspark.sql.functions import desc , asc
df_pyspark.groupBy('Department').max().sort(asc('max(Salary)')).show()
```

```

from pyspark.sql.functions import sum, col, desc
df_pyspark.groupBy("Department") \
  .agg(sum("salary").alias("sum_salary")) \
  .filter(col("sum_salary") > 40000)  \
  .sort(desc("sum_salary")) \
  .show()

#+-----+----------+
#|state|sum_salary|
#+-----+----------+
#|   NY|    252000|
#|   CA|    171000|
#|   NV|    166000|
#+-----+----------+
```

```
## using filter and sorting function with group by
from pyspark.sql.functions import *
df_pyspark.groupBy('Department') \
          .agg(sum('Salary') \
          .alias('Sum_Salary')) \
          .filter(col('Sum_Salary')>40000) \
          .sort(desc('Sum_Salary')).show() 
```

```
## dividing the above query into parts
df_groupby=df_pyspark.groupBy('Department').agg(sum('Salary').alias('Sum_Salary'))

df_filter=df_groupby.filter(col('Sum_Salary')>40000)

df_filter.sort('Sum_salary').show()
```

```
## using two filter conditions in the same query

df_groupBy =df_pyspark.groupBy('Department').agg(sum('Salary').alias('Sum_Salary'))  ## group by and renaming the column
df_filter=df_groupBy.filter((col('Sum_Salary')>30000) &(col('Sum_Salary')<60000))    ## applying the filter

df_filter.sort(desc('Sum_Salary')).show() ## sorting and applying the filter
```
