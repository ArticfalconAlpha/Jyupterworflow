```
file='/FileStore/tables/Salary.csv'
df_pyspark=spark.read.csv(file,header=True,inferSchema=True)
```

```
##Filter operations
df_pyspark.filter("Salary>=20000").show()
```

```
##selecting particular columns based on the search result
df_pyspark.filter('Salary<=20000').select('Name', 'Age').show()
```

```
## using two conditions and(&) condition
df_pyspark.filter((df_pyspark['Salary']<20000)&
                  (df_pyspark['Salary']>15000)).show()

```

```
## Or(|) condiiton
df_pyspark.filter((df_pyspark['Salary']>15000) | 
                  (df_pyspark['Salary']<18000)).show()
```

```
## using not operation
df_pyspark.filter(~(df_pyspark['Salary']>=20000)).show()
```
